# Техническое задание

Металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали.Чтобы оптимизировать производственные расходы. Для этого комбинату нужно контролировать температуру сплава.

# Вводные данные от Заказчика

3. Целевой метрикой, по которой будет оцениваться качество модели является MAE (Mean Absolute Error). На тестовой выборке значение MAE необходимо получить не более 6,8.

Для того, что бы более полно понимать задачу, приведем описание этапа обработки стали, указанное в техничком задании.


# Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

```python
def missing_values(df):
    percent_missing = round(df.isnull().sum() * 100 / len(df))
    percent_missing = percent_missing.astype(int)
    missing_value_df = pd.DataFrame({'column_name': df.columns, 'percent_missing': percent_missing}).sort_values(by='percent_missing')
    return missing_value_df
```

## Предобработка данных

### Предобработка данных об электродах

```python
#Приведение наименований столбцов к общепринятому

data_arc_column = {
    'Начало нагрева дугой' : 'heating_start',
    'Конец нагрева дугой' : 'heating_finish',
    'Активная мощность' : 'active_power',
    'Реактивная мощность' : 'reactive_power'
}

data_arc = data_arc.rename(columns=data_arc_column)
```

```python
#Приведение типов данных в столбцах со временем

data_arc['heating_start'] = pd.to_datetime(data_arc['heating_start'], format='%Y-%m-%d %H:%M:%S')
data_arc['heating_finish'] = pd.to_datetime(data_arc['heating_finish'], format='%Y-%m-%d %H:%M:%S')
```

```python
#Рассчитаем время нагрева

data_arc['heating_time'] = (data_arc['heating_finish'] - data_arc['heating_start']).dt.seconds
```

```python
#Расчитаем количество итераций

heating_iteration = data_arc['key'].value_counts().reset_index(name='heating_iteration').rename(columns={'index': 'key'})
heating_iteration.head()
```

Обработка выбросов

```python
#Построим "ящики с усами" для значений с активной и реактивной мощностями

sns.boxplot(data=data_arc[['active_power', 'reactive_power']], palette='magma', width=0.8)
plt.title('Распределение мощностей', fontsize=14)
plt.xlabel('вид мощности')
plt.ylabel('значение')
plt.show()
```

Как еще было видно из описательной статистики в данных о реактивной мощности присутствует выброс на -700, сделаем срез от 0, так как подаваемая мощность отрицательной быть не может

```python
data_arc = data_arc.query('reactive_power > 0')
```

По значениям активной и реактивной мощности дополнительно рассчитаем значение полной мощности по формуле:

$$
S = \sqrt{P^2 + Q^2}
$$

где S - полная мощность, P - активная мощность, Q - реактивная мощноть

```python
data_arc['real_power'] = (data_arc['active_power']**2 + data_arc['reactive_power']**2)**0.5
```

```python
#Построим "ящики с усами" для значений с активной, реактивной и полной мощностями после среза

sns.boxplot(data=data_arc[['active_power', 'reactive_power', 'real_power']], palette='magma', width=0.8)
plt.title('Распределение мощностей', fontsize=14)
plt.xlabel('вид мощности')
plt.ylabel('значение')
plt.show()
```

```python
#Построим также гистограммы 

plt.figure(figsize=(15, 5))
data_arc['active_power'].plot(kind='hist', bins=100, legend=True, color='#0081AF')
data_arc['reactive_power'].plot(kind='hist', bins=100, legend=True, color='#2DC7FF', alpha=0.8)
data_arc['real_power'].plot(kind='hist', bins=100, legend=True, color='#EABA6B', alpha=0.8)
plt.title('Распределение мощностей', fontsize=14)
plt.xlabel('Значение мощности')
plt.show()
```

Как видно из полученных графиков, данные о мощности распределены близко к нормальному распределению, есть небольшой "хвост", который можно считать выбросом, но до объединения таблиц данный выброс обрабатывать не будем

## Подготовка датасета к обучению и обучение

### Разделение датасета на обучающую и тренировачную выборки

```python
# Выделим признаки и целевой признак
features = data.drop(columns='finish_temp', axis=1)
target = data['finish_temp']
print(features.shape, target.shape)
```

```python
# Разделим данные на обучающую и тестовую выборки

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=random_state, shuffle=True
)

print(features_train.shape, features_test.shape, target_train.shape, target_test.shape)
```

Требование Закзачика по количеству строк в обучащей выборке соблюдено.

### Изучение моделей машинного обучения

В качестве исследуемых моделей выбрали:

- Модель CatBoostRegressor;
- Модель LGBMRegressor;
- Модель RandomForestRegressor

По метрике MAE определим лучшую модель. МАЕ должна быть не более 6,8

#### Модель CatBoostRegressor

```python
%%time

pipe = Pipeline([('cbr', CatBoostRegressor(random_state=random_state, verbose=False))])
parameters = {
    'cbr__iterations': [100, 200, 300, 400, 500],
    'cbr__depth' : [3, 5, 7, 10],
    'cbr__learning_rate' : [0.01, 0.05, 0.1]}

cbr_grid = RandomizedSearchCV(pipe,
                              param_distributions=parameters,
                              scoring='neg_mean_absolute_error',
                              verbose=5,
                              n_jobs=-1) 
cbr_grid.fit(features_train, target_train) 

cbr_MAE_train = round(abs(cbr_grid.best_score_), 3)
cbr_best_params = cbr_grid.best_params_

print()
print(f"Значение MAE модели на трейне: {cbr_MAE_train}")
print(f"Оптимальные гиперпараметры для модели: {cbr_grid.best_params_}")
print()
```

#### Модель LGBMRegressor

```python
%%time

pipe = Pipeline([('lgbm', LGBMRegressor(random_state=random_state, force_col_wise=True))])
parameters = {
    'lgbm__max_depth': [5, 8, 10],
    'lgbm__num_leaves' : [10, 50, 100],
    'lgbm__learning_rate' : [0.01, 0.05, 0.1]}

lgbm_grid = RandomizedSearchCV(pipe,
                              param_distributions=parameters,
                              scoring='neg_mean_absolute_error',
                              verbose=5,
                              n_jobs=-1) 
lgbm_grid.fit(features_train, target_train) 

lgbm_MAE_train = round(abs(lgbm_grid.best_score_), 3)
lgbm_best_params = lgbm_grid.best_params_

print()
print(f"Значение MAE модели на трейне: {lgbm_MAE_train}")
print(f"Оптимальные гиперпараметры для модели: {lgbm_grid.best_params_}")
print()
```

#### Модель RandomForestRegressor

```python
%%time

pipe = Pipeline([('rfr', RandomForestRegressor(random_state=random_state))])
parameters = {
    'rfr__n_estimators': [10, 50, 100],
    'rfr__max_depth' : [5, 10, 15],
    'rfr__min_samples_split' : [2, 4, 8]}

rfr_grid = RandomizedSearchCV(pipe,
                              param_distributions=parameters,
                              scoring='neg_mean_absolute_error',
                              verbose=5,
                              n_jobs=-1) 
rfr_grid.fit(features_train, target_train) 

rfr_MAE_train = round(abs(rfr_grid.best_score_), 3)
rfr_best_params = rfr_grid.best_params_

print()
print(f"Значение MAE модели на трейне: {rfr_MAE_train}")
print(f"Оптимальные гиперпараметры для модели: {rfr_grid.best_params_}")
print()
```

#### Выводы по обучению

Обучили три модели: 
- Модель CatBoostRegressor. Значение МАЕ на обучающей выборке = 5.934;
- Модель LGBMRegressor. Значение МАЕ на обучающей выборке = 6.007;
- Модель RandomForestRegressor. Значение МАЕ на обучающей выборке = 6.091.

Хоть и не с большим разрывом, но самой оптимальной моделью по полученным значениям будем считать модель CatBoostRegressor. Ее и будем проверять на тестовой выборке

Так как наибольшей важностью признака обладает первая измеренная температура, то при выполнении технологического процесса рекомендуется более тщательный ее контроль. Также рекомендуется контролировать время нагрева ковша дугой, так как при перегреве количество затрачиваемой электроэнергии может увеличиваться. Можно также контролировать время всего процесса (от первого до последнего измерения). Например, можно доработать технологический процесс и установить допустимые для выполнения всех требуемых итераций временные "рамки", таким образом можно тоже уменьшить траты на электроэнергию за счет уменьшения времени включения и поддержания во включенном состоянии оборудования.

Так же есть вероятность улучшить полученную модель, в случае определения заказчиком пороговых значений не только по температуре, но и по другим признакам технологического процесса, для удобства обработки выбросов.
